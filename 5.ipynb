# 1) Виконання вправ по 5-му варіанту (3, 5, 7, 9, 10, 12, 14, 15, 21)

```python
def timer(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start = time.perf_counter()
        func(*args, **kwargs)
        end = time.perf_counter()
        print(f"Було виконано за {end - start:.6f} секунд.")
    return wrapper
    

@timer
def task_3(n, m, r, c):
    A = np.ones((n, m))
    for i in range(n):
        for j in range(m):
            if i % r == 0 or j % c == 0:
                A[i, j] = 0
    print("3 завдання з бібліотекою")
    print(A)


@timer
def task3(n, m, r, c):
    A = [[1 for x in range(m)] for y in range(n)]
    for i in range(n):
        for j in range(m):
            if i % r == 0 or j % c == 0:
                A[i][j] = 0
    print("3 завдання без бібліотеки")
    for row in A:
        print(*row)


@timer
def task_5(n):
    A = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            if i % 2 == 0:
                A[i, j] = 1
    print("5 завдання з бібліотекою")
    print(A)


@timer
def task5(n):
    A = [[0 for x in range(n)] for y in range(n)]
    for i in range(n):
        for j in range(n):
            if i % 2 == 0:
                A[i][j] = 1
    print("5 завдання без бібліотеки")
    for row in A:
        print(*row)


@timer
def task_7(*args):
    A = np.array(args, dtype=float)
    A[A == 0] = -1
    print("7 завдання з бібліотекою")
    print(A)


@timer
def task7(*args):
    A = list(args)
    A = [-1 if x == 0 else x for x in A]
    print("7 завдання без бібліотеки")
    print(A)


@timer
def task_9(n):
    A = np.zeros(n+1)
    for i in range(n):
        A[i] = n - i
    print("9 завдання з бібліотекою")
    print(A)


@timer
def task9(n):
    A = list()
    for i in range(n+1):
        A.append(n-i)
    print("9 завдання без бібліотеки")
    print(A)


@timer
def task_10(n, m):
    A = np.ones((n, m))
    for i in range(n):
        for j in range(m):
            A[i, j] = random.randint(-10, 10)
    print("10 завдання з бібліотекою")
    print(f"мінімум: {np.min(A):.3f}")
    print(f"максимум: {np.max(A):.3f}")
    print(f"середнє: {np.mean(A):.3f}")
    print(f"дисперсія: {np.var(A):.3f}")


@timer
def task10(n, m):
    A = [[random.randint(-10, 10) for x in range(m)] for x in range(n)]
    AA = [x for row in A for x in row]
    print("10 завдання без бібліотеки")
    print(f"мінімум: {min(AA):.3f}")
    print(f"максимум: {max(AA):.3f}")
    print(f"середнє: {(sum(AA) / len(AA)):.3f}")
    print(f"дисперсія: {(sum((x - (sum(AA) / len(AA))) ** 2 for x in AA) / len(AA)):.3f}")


@timer
def task_12(n):
    A = np.ones((n, n))
    for i in range(n):
        for j in range(n):
            if i == 0 or j == 0 or i == n-1 or j == n-1:
                A[i, j] = 0
    print("12 завдання з бібліотекою")
    print(A)


@timer
def task12(n):
    A = [[1 for x in range(n)] for y in range(n)]
    for i in range(n):
        for j in range(n):
            if i == 0 or j == 0 or i == n-1 or j == n-1:
                A[i][j] = 0
    print("12 завдання без бібліотеки")
    for row in A:
        print(*row)


@timer
def task_14():
    pattern = np.array([[0, 1],
                    [1, 0]])
    A = np.tile(pattern, (4, 4))
    print("12 завдання з бібліотекою")
    print(A)


@timer
def task_15(n):
    A = np.ones((n, n))
    for i in range(n):
        for j in range(n):
            if j % 2 == 0:
                A[i, j] = 0
    print("15 завдання з бібліотекою")
    print(A)


@timer
def task15(n):
    A = [[1 for x in range(n)] for y in range(n)]
    for i in range(n):
        for j in range(n):
            if j % 2 == 0:
                A[i][j] = 0
    print("15 завдання без бібліотеки")
    for row in A:
        print(*row)


@timer
def task_21(n, m):
    A = np.ones((n, n))
    for i in range(n):
        for j in range(n):
            A[i, j] = random.randint(0, m)
    B = np.max(A)
    for i in range(n):
        for j in range(n):
            if A[i, j] == B:
                A[i, j] *= -1
    print("21 завдання з бібліотекою")
    print(A)


@timer
def task21(n, m):
    A = [[random.randint(0, m) for x in range(n)] for y in range(n)]
    AA = [x for row in A for x in row]
    B = max(AA)
    for i in range(n):
        for j in range(n):
            if A[i][j] == B:
                A[i][j] *= -1
    print("21 завдання без бібліотеки")
    for row in A:
        print(*row)
```

# 2) Метод Крамера і перевірки

```python
def cramer_method(A, b):
    det_A = np.linalg.det(A)  # Детермінант
    if abs(det_A) < 1e-10:
        raise ValueError("Система не має єдиного розв'язку (det(A) = 0).")

    n = len(b)
    x = np.zeros(n)
    print("Крамер: ")
    for i in range(n):
        A_i = np.copy(A)  # Копія
        A_i[:, i] = b  # Заміна значень для копії
        det_Ai = np.linalg.det(A_i)  # Детермінант копії
        x[i] = det_Ai / det_A  # Розрахунок
        print(A_i)
        print(x[i])
        print()
    return x


def check_method1(A, b, x):  # Перевірка
    check_b = A @ x  # Матричне множення
    return np.allclose(check_b, b)


def check_method2(A, b, x):  # Перевірка
    x_invert = np.linalg.inv(A) @ b  # Обернена матриця
    return np.allclose(x_invert, x)


def check_method3(A, b, x):  # Перевірка
    x_solve = np.linalg.solve(A, b)  # Функція розрахунку
    return np.allclose(x_solve, x)
```

```python
A = np.array([
    [1, 3, 5, 7],
    [3,  5, 7, 1],
    [5,  7,  1, 3],
    [7, 1,  3,  5]
], dtype=float)

b = np.array([12, 0, 4, 16], dtype=float)

x = cramer_method(A, b)
print("Перевірка результатів")
print(check_method1(A, b, x))
print(check_method2(A, b, x))
print(check_method3(A, b, x))
```

# 3) Обчислення матричного виразу

```python
def mat_mat(X, Y):
    n, m, p = len(X), len(Y), len(Y[0])
    result = [[0 for _ in range(p)] for _ in range(n)]
    for i in range(n):
        for j in range(p):
            for k in range(m):
                result[i][j] += X[i][k] * Y[k][j]
    return result


def add(X, Y):
    return [[X[i][j] + Y[i][j] for j in range(len(X[0]))] for i in range(len(X))]


def sub(X, Y):
    return [[X[i][j] - Y[i][j] for j in range(len(X[0]))] for i in range(len(X))]


def mat_num(X, k):
    return [[X[i][j] * k for j in range(len(X[0]))] for i in range(len(X))]


@timer
def C_numpy(A, B):
    C = (A - B @ B) @ (2 * A + B)
    print(C)


@timer
def CC_no_numpy(AA, BB):
    CC1 = sub(AA, mat_mat(BB, BB))
    CC2 = add(mat_num(AA, 2), BB)
    CC = mat_mat(CC1, CC2) 
    for row in CC:
        print(row)
```

```python
A = np.array([[5, 2, 0], 
              [10, 4, 1], 
              [7, 3, 2]])

B = np.array([[3, 6, -1], 
              [-1, -2, 0], 
              [2, 1, 3]])

AA = [[5, 2, 0],
     [10, 4, 1],
     [7, 3, 2]]

BB = [[3, 6, -1],
     [-1, -2, 0],
     [2, 1, 3]]

print("3) завдання з бібліотекою:")
C_numpy(A, B)
print()
print("3) завдання без бібліотеки:")
CC_no_numpy(AA, BB)
```
